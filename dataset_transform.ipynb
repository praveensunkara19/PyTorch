````python
import torch
import torchvision
from torch.utils.data import Dataset
import numpy as np
````

````python
class WineDataset(Dataset):

    def __init__(self, transform = None):
        #data loading
        xy = np.loadtxt('data\\wine.csv', delimiter=',', dtype=np.float32, skiprows=1)
        self.x = xy[: , 1:]  #features of all rows & colums
        self.y = xy[: , [0]] # target of all rows and ist column

        self.transform = transform


    def __getitem__(self, index):
        #dataset[0]
        sample = self.x[index], self.y[index]

        if self.transform:
            sample = self.transform(sample)

        return sample

    def __len__(self):
        #len(dataset)
        return self.n_samples

# custom Tranforms

class ToTensor:
    def __call__(self, sample):
        inputs, targets = sample
        return torch.from_numpy(inputs), torch.from_numpy(targets)
    
class MulTransform:
    def __init__(self, factor):
        self.factor = factor

    def __call__(self, sample):
        inputs, targets = sample
        inputs *= self.factor
        return inputs, targets


````

````python
dataset = WineDataset(transform=None)
first_data = dataset[0]
print(first_data)
print(type(first_data))

features, lables = first_data
print(type(features), type(lables))
````

````python
# usign the compose to applying the ToTensor and MulTransform clases

composed = torchvision.transforms.Compose(transforms= [ToTensor(), MulTransform(2)])
````

````python
dataset = WineDataset(transform=composed)
first_data = dataset[0]
print(first_data)
print(type(first_data))

features, lables = first_data
print(type(features), type(lables))
````
