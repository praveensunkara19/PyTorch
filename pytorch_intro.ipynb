```python
import torch

x = torch.rand(2,2)
y = torch.rand(2,2)

z = x + y

print(z)

z = torch.add(x,y)
print(z)

y.add_(x)
print(y)

print(z == y)
```

```python
a = torch.rand(2,2)
b = torch.rand(2,2)

c = a - b
print(c)

c = torch.sub(a,b)
print(c)

c = a.sub_(b)
print(c)


"""
same way for div, mul
"""
```

Gradients
* needed for model optimization

```python
# autograd

import torch

x = torch.rand(3, requires_grad=True)
print('x-ten',x)

y = x+2 
print('y-ten',y)

z = y*y*2
# z = z.mean()
print('z-ten', z)

v = torch.tensor([0.1, 1.0, 0.001], dtype=float)
z.backward(v) #dz/dx. # this only works for the scaler values i.e z.mean() or if z is vector, we neet to pass other vec to make it a dot product(joccovian dot product)
print(x.grad)
```

```python
x = torch.randn(3, requires_grad=True)
print(x)

# x.requires_grad_(False) #with this functino we can make a tensor not to use grad any time
# x.detach()
#this allows not to use grad() fnc
with torch.no_grad():
    y = x+2
    print(y)

print('with gradient')
y = x + 2
print(y)
```

```python
# During the training the backword() function passes the grad values of the iteration
# So, we need to make the grad func empty for every iteration, make sure it doesnot carry the previous grad values

weights = torch.ones(4, requires_grad=True)

for epoch in range(3):
    model_output = (weights*3).sum()
    
    model_output.backward()

    print(weights.grad)

    weights.grad.zero_() #this stops the previous iterations grad values that effects the current ones
```
