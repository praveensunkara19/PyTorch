Activation Function: apply to non-linear transformation ad decide whether a neuron should be activated or not!

### Most popular activation funtions:

1. Step function f(x) = 1 if x>=0 else 0 otherwise
2. Sigmoid       f(x) = 1/ 1 + e^-x 
3. TanH
4. ReLu
5. Leaky ReLu
6. Softmax

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

Option 1

```python
# defining the activation fun in Neural Nets

class NeuralNet(nn.Module):
    def __init__(self, input_size, hidden_size):
        super(NeuralNet, self).__init__()
        self.linear = nn.Linear(input_size, hidden_size) #input and ouput features
        self.relu = nn.ReLU()
        self.linear2 = nn.Linear(hidden_size, 1) #output is always single values i.e linear func
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        out = self.linear(x)
        out = self.relu(out)
        out = self.linear2(out)
        out = self.sigmoid(out)

        return out
```

option 2

```python
# calling the activation fun in forward layers form nn module or nn.functional


class NeuralNet(nn.Module):
    def __init__(self, input_size, hidden_size):
        super(NeuralNet, self).__init__()
        self.linear1 = nn.Linear(input_size, hidden_size)
        self.linear2 = nn.Linear(hidden_size, 1)

       
    def forward(self, x):
        out = nn.ReLU(self.linear1(x))
        out = nn.Sigmoid(self.linear2(out))
        # out = F.sigmoid()
        # out = F.leaky_relu()
        return out
```
