# Building the Graph

```python
# <!--  z = x*y , y = a + b -->

a = torch.tensor([4.0], requires_grad=True)
b = torch.tensor([5.0], requires_grad=True)
x = torch.tensor([6.0], requires_grad=True)


y = a + b

z = x * y

print(y)
print(z)
```

```python
print(f"grad_fn for a is {a.grad_fn}")   #coz in defines as a variable not funtion

print(f"grad_fn for y is {y.grad_fn}")
print(f"grad_fn for z is {z.grad_fn}")
```

### Element-wise vs Matrix multiplicatin

Element wise Multiplication
* Multiplies mathces position
* tensors must have same shape 

```python
a = torch.tensor([[1,2],[3,4]])
b = torch.tensor([[10,20],[30,40]])

mul = a * b  #1*10, 2*20, 3*30, 4*40

mul
```

Matrix multiplication powers the newral networks

* m1 columns = m2 rows

```python
matrix_mul = a @ b
matrix_mul
```

```python
m1 = torch.tensor([[1,2,3],[4,5,6]])  #2 X 3
m2 = torch.tensor([[7,8],[9,10],[11,12]])  # 3 X 2

print(m1@m2)
```

 to write the linear equation 

* y = mx + c represeted as y = m @ x + c

### Reduction Operators with dim arguments

```python
scores = torch.tensor([[10.,20.,30.],[5.,6.,7.,]])  # 2 X 3

avg_per_assign = scores.mean(dim=0)  # columns
avg_per_stud   = scores.mean(dim=1)  # rows

print(avg_per_assign)
print(avg_per_stud)
```

* argmax

```python
scores = torch.tensor([[10, 0, 5, 20, 1],
                       [1, 30, 2, 5, 0]])

best_indices = torch.argmax(scores, dim=1)  #max values of each row

print(best_indices)
```

torch.gather()

```python
data = torch.tensor([
    [10, 11, 12, 13],
    [20, 21, 22, 23],
    [30, 31, 32, 33]
])

indices_to_select = torch.tensor([[2], [0], [3]])

selected_values = torch.gather(data, dim=1, index=indices_to_select)

selected_values

# tensor([[12],  for row 0, it got indext 2
#         [20],  for row 1, it got indext 0
#         [33]]) for row 2, it got indext 3
```
