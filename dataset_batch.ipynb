### Dataloder, Datasets, Batchs
    '''
    epoch -> 1 forward and backward pass of ALL Training Samples
    batch_size -> number of training samples in one forward & backward pass
    number of iterations -> number of passes, each pass using [batch_size] no.of samples
    eg: 100 samples, batch_size = 20 -> 100/20 = 5 iterations for 1 epoch
'''

```python
import torch
import torchvision
from torch.utils.data import Dataset, DataLoader
import numpy as np
import math
```

```python
class WineDataset(Dataset):

    def __init__(self):
        #data loading
        xy = np.loadtxt('data\\wine.csv', delimiter=',', dtype=np.float32, skiprows=1)
        self.x = torch.from_numpy(xy[: , 1:])  #features of all rows & colums
        self.y = torch.from_numpy(xy[: , [0]]) # target of all rows and ist column
        self.n_samples = xy.shape[0]

    def __getitem__(self, index):
        #dataset[0]
        return self.x[index], self.y[index]

    def __len__(self):
        #len(dataset)
        return self.n_samples


# looking into the 1st row of the dataset

dataset = WineDataset()
first_data = dataset[0]
print(first_data)
# features, labels = first_data
# print(features, labels)
```

```python
print(dataset.__len__())
print(len(dataset))
```

```python
# DataLoader -- used to load the data in the batch_size

dataloader = DataLoader(dataset=dataset, batch_size=4, shuffle=True)

dataiter = iter(dataloader)
data = dataiter._next_data()
features, data = data
print(features, data)

```

```python
# training loop

num_epochs = 2
total_smaples = len(dataset)
n_iterations = math.ceil(total_smaples/4)
print(total_smaples, n_iterations)
```

```python
for epoch in range(num_epochs):
    for i, (input, labels) in enumerate(dataloader):
        #forward backward, update
        if (i+1) % 5 == 0:
            print(f'epoch {epoch+1}/{num_epochs}, step {i+1}/{n_iterations}, inputs {input.shape}')
```
